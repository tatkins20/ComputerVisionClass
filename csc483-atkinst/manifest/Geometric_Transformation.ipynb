```python
import cv2
import math
import matplotlib.pyplot as plt
import numpy as np
from geotransform import *

img = cv2.imread('../images/project3/demo.pgm')
plt.imshow(img)
plt.show()
                            
```

For the affine transformation function I use Open CV's get affine transform function to get the transformation matrix. I then apply this transformation matrix to an array that contains every single (x,y) point in the image. After I multiply the matrix, I create a image that the new transformed image can map onto.

```python
rows, cols, ch = img.shape
in_pts = np.float32([[10,50],[110,200],[450,100]])
out_pts = np.float32([[10,50],[100,50],[100,200]])
matrix = cv2.getAffineTransform(in_pts, out_pts)
affineImg = AffineTransform(img, matrix)
plt.imshow(affineImg, cmap='gray')
plt.show()
```
